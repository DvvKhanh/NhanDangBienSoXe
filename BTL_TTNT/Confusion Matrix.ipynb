```python
# =============================
# 1. IMPORT THƯ VIỆN
# =============================
import pandas as pd
import matplotlib.pyplot as plt

# =============================
# 2. TẠO BẢNG SỐ LIỆU
# =============================
data = {
    "Dataset": ["Train", "Validation", "Test"],
    "Accuracy (%)": [98.5, 96.8, 96.3],
    "Loss": [0.05, 0.12, 0.14]
}

df = pd.DataFrame(data)

# Hiển thị bảng
df
```

```python
plt.figure()
bars = plt.bar(df["Dataset"], df["Accuracy (%)"])

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, round(yval,2),
             ha='center', va='bottom')

plt.title("So sánh Accuracy giữa các tập dữ liệu")
plt.ylabel("Accuracy (%)")
plt.ylim(90, 100)
plt.show()
```

```python
plt.figure()
bars = plt.bar(df["Dataset"], df["Loss"])

for bar in bars:
    yval = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        yval + 0.005,
        round(yval, 3),
        ha='center',
        va='bottom'
    )

plt.title("So sánh Loss giữa các tập dữ liệu")
plt.ylabel("Loss")
plt.ylim(0, 0.16)
plt.show()
```
